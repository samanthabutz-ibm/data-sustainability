```python
def parse_table(soup):
    
    try:
        my_table = soup.find('table',{'class':'infobox vcard'}).findAll('tr')
    except:
        return None
    
    data = {}
    for tr in my_table:
        # print(tr)

        header = tr.find('th',{'class':'infobox-label'})
        content = tr.find('td',{'class':'infobox-data'})

        try:
            # print(header.text)
            # print(content.text)

            data[header.text] = content.text
            # print()
        except:
            pass
        
    return data
```
